```python
import pandas as pd
import sqlite3

print("Libraries Imported Successfully!")
```

```python
df = pd.read_csv("cleaned_data.csv")

df.head()
```

```python
conn = sqlite3.connect("apple_products.db")
```

```python
df.to_sql(
    "AppleProducts",
    conn,
    if_exists="replace",
    index=False
)

print("Table created successfully!")
```

```python
query = """
SELECT *
FROM AppleProducts
LIMIT 5;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT DISTINCT Platform
FROM AppleProducts;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT COUNT(DISTINCT Platform) AS Total_Platforms
FROM AppleProducts;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    Platform,
    AVG(Discount_Pct) AS Average_Discount
FROM AppleProducts
GROUP BY Platform
ORDER BY Average_Discount DESC;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    Model_Name,
    Platform,
    Product_Category,
    Rating
FROM AppleProducts
WHERE Rating >= 4.8
ORDER BY Rating DESC;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    Model_Name,
    Product_Category,
    Platform,
    Current_Price_USD
FROM AppleProducts
WHERE Product_Category = 'iPhone';
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    Model_Name,
    Platform,
    Stock_Status
FROM AppleProducts
WHERE Stock_Status = 'Out of Stock';
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    Model_Name,
    Platform,
    Rating
FROM AppleProducts
WHERE Product_Category = 'iPhone'
AND Rating > 4.7;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    Model_Name,
    Discount_Pct,
    CASE
        WHEN Discount_Pct >= 30 THEN 'High Discount'
        WHEN Discount_Pct >= 15 THEN 'Medium Discount'
        ELSE 'Low Discount'
    END AS Discount_Category
FROM AppleProducts
LIMIT 20;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    CASE
        WHEN Discount_Pct >= 30 THEN 'High Discount'
        WHEN Discount_Pct >= 15 THEN 'Medium Discount'
        ELSE 'Low Discount'
    END AS Discount_Category,

    AVG(Rating) AS Average_Rating

FROM AppleProducts

GROUP BY Discount_Category

ORDER BY Average_Rating DESC;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    Product_Category,
    Model_Name,
    Rating,

    RANK() OVER(
        PARTITION BY Product_Category
        ORDER BY Rating DESC
    ) AS Product_Rank

FROM AppleProducts;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
WITH RankedProducts AS (

SELECT
    Product_Category,
    Model_Name,
    Rating,

    RANK() OVER(
        PARTITION BY Product_Category
        ORDER BY Rating DESC
    ) AS Product_Rank

FROM AppleProducts

)

SELECT *

FROM RankedProducts

WHERE Product_Rank = 1;
"""

pd.read_sql_query(query, conn)
```
